Назначающей возрастной рейтинг для игр в Штатах является организация ESRB, всего есть 7 вариантов рейтинга, куда не входит `K-A` в данный момент.
По информации из Интернета, сначала было всего 5 вариантов рейтинга, где и было такое наименование:

K-A: Для детей – для тех, кто уже научился говорить и держать джойстик, эти игры уже могут вызывать интерес [позже в 1998 году был переименован в Е – для всех возрастов, а также чуть позже был добавлен рейтинг Е10+, как промежуточный между Е и Т]


Поэтому можно заменить его на значение E.

```python
df_game['rating'] = df_game['rating'].replace('K-A', 'E')
```

```python
print(df_game['rating'].value_counts())
print(df_game['rating'].unique())
```

```python
display(df_game[df_game['rating'].isna()].sample(10))
```

В столбце 6764 пропуска из 16713, то есть 40,47%. Почти половина. Эти данные можно было бы восстановить, если бы у нас было больше данных или меньше пропусков в зависимоси от жанра игры.

Также при анализе влияние рейтинга на продажи в зависимости от региона наши усредненные значения будут выбиваться, так как помимо жанра все же есть разные условия установки рейтинга. Информация из интернета:

"Оценка проводится по нескольким основным позициям. Эксперты раскладывают игру на элементы, которые учитываются в рейтинге, и сверяют их со списком. За каждый элемент начисляется определенное число штрафных очков. Результаты плюсуются, и в зависимости от получившейся суммы выставляется итоговый рейтинг." 
Всего 22 критерия.

Поэтому, как в случае с годом, лучше заменить пропуски целочисленным значением и при построении графиков делать срез без учета этих значений.

## Проверка гипотез

Необходимо проверить 2 гипотезы:

1. Средние пользовательские рейтинги платформ Xbox One и PC одинаковые;

2. Средние пользовательские рейтинги жанров Action (англ. «действие», экшен-игры) и Sports (англ. «спортивные соревнования») разные.

Рассчитаем средние пользовательские оценки для платформ Xbox One и PC и построим графики:

```python
display(df_game_actual.query('platform == "XOne"')['user_score'].describe())
df_game_actual.query('platform == "XOne"')['user_score'].plot(kind='hist',grid=True, figsize=(7, 5), color='#007fff')
plt.title('График пользовательского рейтинга платформы Xbox One')
# '#007fff' лазурный
```

```python
display(df_game_actual.query('platform == "PC"')['user_score'].describe())
df_game_actual.query('platform == "PC"')['user_score'].plot(kind='hist',grid=True, figsize=(7, 5), color='#007fff')
plt.title('График пользовательского рейтинга платформы PC')
```

Проверка гипотез начинается с формулировки нулевой гипотезы H₀:
    
**H₀** = Средние пользовательские рейтинги платформ Xbox One и PC одинаковые;

Исходя из H₀ формулируется альтернативная гипотеза H₁. Это утверждение, которое принимается верным, если отбрасывается H₀:

**H₁** = Средние пользовательские рейтинги платформ Xbox One и PC различны.

Нам необходимо сравнить 2 генеральные совокупности, поэтому воспользуемся методом scipy.stats.ttest_ind (наши генеральные совокупности не зависят друг от друга и средние нормально распределены), которому передадим таблицы с разными платформами:

```python
alpha = .01 # критический уровень статистической значимости

results = st.ttest_ind(
    df_game_actual.query('platform == "XOne"')['user_score'].dropna(), 
    df_game_actual.query('platform == "PC"')['user_score'].dropna())

print('p-значение:', results.pvalue)

if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу") 
```

Исходя из графиков и расчетов, а также проверки гипотез, можно сказать, что средние пользовательские рейтинги платформ Xbox One (6,52) и PC (6,27) одинаковые.

Расчитаем средние пользовательские оценки для жанров Action и Sports и построим графики:

```python
display(df_game_actual.query('genre == "Action"')['user_score'].describe())
df_game_actual.query('genre == "Action"')['user_score'].plot(kind='hist',grid=True, figsize=(7, 5), color='#ff8800')
plt.title('График пользовательского рейтинга жанра Action')
# '#ff8800' - мандариновый
```

```python
display(df_game_actual.query('genre == "Sports"')['user_score'].describe())
df_game_actual.query('genre == "Sports"')['user_score'].plot(kind='hist',grid=True, figsize=(7, 5), color='#ff8800')
plt.title('График пользовательского рейтинга жанра Sports')
```

Проверим **2-ю** гипотезу:
    
**H₀** = Средние пользовательские рейтинги жанров Action и Sports не отличаются.

**H₁** = Средние пользовательские рейтинги жанров Action и Sports разные.

```python
alpha = .01 # критический уровень статистической значимости

results = st.ttest_ind(
    df_game_actual.query('genre == "Action"')['user_score'].dropna(), 
    df_game_actual.query('genre == "Sports"')['user_score'].dropna())

print('p-значение:', results.pvalue)

if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

Исходя из графиков и расчетов, а также проверки гипотез, можно сказать, что средние пользовательские рейтинги жанров Action (среднее 6,84) и Sports (среднее 5,24) разные.
