# Исследование объявлений о продаже квартир

В вашем распоряжении данные сервиса Яндекс.Недвижимость — архив объявлений о продаже квартир в Санкт-Петербурге и соседних населённых пунктах за несколько лет. Нужно научиться определять рыночную стоимость объектов недвижимости. Ваша задача — установить параметры. Это позволит построить автоматизированную систему: она отследит аномалии и мошенническую деятельность. 

По каждой квартире на продажу доступны два вида данных. Первые вписаны пользователем, вторые получены автоматически на основе картографических данных. Например, расстояние до центра, аэропорта, ближайшего парка и водоёма. 

**Описание данных:**
 * `airports_nearest` — расстояние до ближайшего аэропорта в метрах (м)
 * `balcony` — число балконов
 * `ceiling_height` — высота потолков (м)
 * `cityCenters_nearest` — расстояние до центра города (м)
 * `days_exposition` — сколько дней было размещено объявление (от публикации до снятия)
 * `first_day_exposition` — дата публикации
 * `floor` — этаж
 * `floors_total` — всего этажей в доме
 * `is_apartment` — апартаменты (булев тип)
 * `kitchen_area` — площадь кухни в квадратных метрах (м²)
 * `last_price` — цена на момент снятия с публикации
 * `living_area` — жилая площадь в квадратных метрах (м²)
 * `locality_name` — название населённого пункта
 * `open_plan` — свободная планировка (булев тип)
 * `parks_around3000` — число парков в радиусе 3 км
 * `parks_nearest` — расстояние до ближайшего парка (м)
 * `ponds_around3000` — число водоёмов в радиусе 3 км
 * `ponds_nearest` — расстояние до ближайшего водоёма (м)
 * `rooms` — число комнат
 * `studio` — квартира-студия (булев тип)
 * `total_area` — площадь квартиры в квадратных метрах (м²)
 * `total_images` — число фотографий квартиры в объявлении

Далее напишем функцию, которая будет проверять значение столбца `'floors_total'`, и, если значение будет нулевым, функция поменяет его на значение столбца `'floor'` (сделаем так из предположения, что количество этажей дома не указано из-за того, что квартира расположена на последнем этаже).

```python
def floors_total(row):
    """
    Проверяет значение столбца 'floors_total',
    и если оно нулевое, заменяет на значение из 
    'floor'
    """
    if row[6] == 0:
        row[6] = row[8]
    return row
```

```python
# Сначала я хотела заменить значения жилой площади и площади кухни в зависимости от количества комнат, 
# но количество квартир, в которых сумма жилой и площади кухни больше общей площади увеличилась.
#re_data['living_area'] = re_data['living_area'].fillna(re_data.groupby('rooms')['living_area'].transform('median'))
#re_data['kitchen_area'] = re_data['kitchen_area'].fillna(re_data.groupby('rooms')['kitchen_area'].transform('median'))
```

```python
# Заменим пропуски нулями
re_data['living_area'] = re_data['living_area'].fillna(0)
re_data['kitchen_area'] = re_data['kitchen_area'].fillna(0)
```

Для того, чтобы заменить нули, нам нужно знать, какое соотношение жилой площади и площади кухни к общей площади:

```python
living_area_to_total_median = (re_data['living_area'] / re_data['total_area']).median()
kitchen_area_to_total_median = (re_data['kitchen_area'] / re_data['total_area']).median()

print(living_area_to_total_median)
print(kitchen_area_to_total_median)
```

```python
def living_and_kitchen_area(row):
    """
    Проверяет значение столбца 'living_area' и 'kitchen_area',
    и если оно нулевое, меньше 5 или общая площадь меньше суммы кухни и жилой площади,
    рассчитывает соотношение жилой площади или площади кухни как 55,7% и 16,2% 
    от общей и перезаписывает.
    """
    if row[7] == 0 or row[7] < 6 or (row[2]<(row[7]+row[12])):
        row[7] = float(round(row[2] * living_area_to_total_median, 2))
    if row[12] == 0 or (row[2]<(row[7]+row[12])):
        row[12] = float(round(row[2] * kitchen_area_to_total_median, 2))   
    return row
```

```python
re_data = re_data.apply(living_and_kitchen_area, axis=1)

re_data['living_area'].isna().value_counts()
```

Осталось рассмотреть столбец `'ceiling_height'`.

Потолки ниже 2,5 м скорее всего из-за того, что есть подвесной потолок или конструкция на потолке, которая была не демонтирована и поэтому указана в объявлении. Потолки от 25 до 32 м скорее всего имеют ошибку на порядок, их можно поделить на 10. 
Потолки от 5 до 6 м имеют квартиры более 100 м2, это может быть реально. 

Все выбивающиеся значения замениим на медианное значение - 2,65 м

```python
re_data['ceiling_height'] = re_data['ceiling_height'].fillna(0)
```

```python
def ceiling_height(row):
    """
    Проверяет значение столбца 'ceiling_height' и если оно меньше 2.4 
    или больше 8, но меньше 23 или равно 100, то заменяет на медианное 2,65м, 
    а значения в диапазоне от 24 до 32 делит на 10.
    """
    if row[5] <= 2.4 or 8 <= row[5] <= 23 or row[5] == 100:
        row[5] = 2.65
    elif 24 <= row[5] <= 32:
        row[5] = row[5]/10 
    return row
```

```python
re_data = re_data.apply(ceiling_height, axis=1)
```

Теперь переведем некоторые столбцы в нужный тип методом `astype()`:

```python
re_data['last_price'] = re_data['last_price'].astype('int') # копейки не указаны
re_data['floors_total'] = re_data['floors_total'].astype('int') # общее количество этажей не может быть не целым
re_data['balcony'] = re_data['balcony'].astype('int') # общее количество балконов не может быть не целым
re_data['airports_nearest'] = re_data['airports_nearest'].astype('int', errors='ignore') # расстояние указано в м, бОльшая точность нам не нужна
re_data['cityCenters_nearest'] = re_data['cityCenters_nearest'].astype('int', errors='ignore') # -//- (как в предыдущем)
re_data['parks_nearest'] = re_data['parks_nearest'].astype('int', errors='ignore') # -//- (как в предыдущем)
re_data['ponds_nearest'] = re_data['ponds_nearest'].astype('int', errors='ignore') # -//- (как в предыдущем)
re_data['parks_around3000'] = re_data['parks_around3000'].astype('int') # число парков не может быть не целым
re_data['ponds_around3000'] = re_data['ponds_around3000'].astype('int') # число водоемов не может быть не целым
re_data['days_exposition'] = re_data['days_exposition'].astype('int', errors='ignore') # дни также целое число
```

Для добавления столбца с этажем квартиры, нужно написать функцию:

```python
def floor_category(row):
    """
    Проверяет значение столбца 'floor', и если оно равно 1, то 
    значение категории этажа - первый, если равно столбцу 'floors_total', то
    это последний этаж, и третья категория для остальных этажей
    """
    if row[8] == 1:
        return 'первый'
    elif row[8] == row[6]:
        return 'последний'
    else:
        return 'другой'
```

## Исследовательский анализ данных

Изучим следующие параметры: площадь, цена, число комнат, высота потолков и построим гистограммы для каждого.

Напишем функцию для построения графиков:

```python
def serie_hist(serie, bins=0):
    '''
    Фунция получает столбец датафрейма, берет значения из метода .describe(),
    рассчитывает по полученным значениям левую и правую границу графика,
    рассчитывает количество bins, передает полученные значения .hist()
    и строит график 
    '''
    # Присваиваем имена параметрам из .describe()
    serie_desc = serie.describe()
    mean = serie_desc[1]
    std = serie_desc[2]
    min_desc = serie_desc[3]
    q25 = serie_desc[4]
    median = serie_desc[5]
    q75 = serie_desc[6]
    max_desc = serie_desc[7]
        
    # Рассчитываем левую и правую границу с учетом перцентилей 
    # и проверяем, не выходяит ли границы за минимальное и максимальное значения
    left = q25 - (q75-q25)*1.5
    right = q75 + (q75-q25)*1.5
    if left < min_desc:
        left = min_desc
    if right > max_desc:
        right = max_desc
     
    # Рассчитываем параметр bins, заменяем в том случае, если параметр не был задан
    # или он получился слишком большим
    if bins == 0:
        bins = right - left
        if bins > 100:
            bins = 100
        elif bins < 1:
            bins = abs(bins*10)
        bins = int(bins)
    else:
        bins = bins
    
    # Строим график
    serie.hist(bins=bins, range=(left, right))
    print(serie_desc)
    return (left, right,bins)
```

В Зеленогорске в среднем, есть квартиры, дороже, чем в Санкт-Петербугре, но проанализировав средние показатели и тот факт, что в Петербурге больше выставлено квартир на продажу, можно сказать, что в центре Санкт-Петербурга самые дорогие квартиры. Также дорогая недвижимость в  Пушкине, Сестрорецке, Кудрово.

В центре города больше средняя площадь квартир (90 кв м., медиана - 76 кв. м.) также связана с тем, что есть объявления о продаже коммунальных квартир с большой площадью (до 900 кв.м).
Средняя цена за кв м 142 тысячи, медианная 123,5 тысячи, квартиры обычно имеют 1-2 комнаты и потолки 3 м (видимо, Сталинская застройка)
По остальным показателям квартиры в центре похожи на квартиры в остальной части города.

Также, хочу заметить, что год постройки дома во многом влияет на стоимость за квадратный метр, и было бы интересно сравнить, окраину города и центр, т.к. обычно на окраине новые дома, а в центре старой застройки, но, к сожалению, такой информации не было в тааблице.
